```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate a simple dataset (y=2x+1 with some added noise)
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 10, 50) # 50 evenly spaced points between 0 and 10
true_m, true_c = 2, 1
y = true_m * x + true_c + np.random.normal(0, 1, x.shape)  # Adding noise
```

```python
#initialize parameters
m, c = 0, 0
learning_rate = 0.01
num_iterations = 1000

#store the loss for plotting
loss_history = []
```

```python
#Gradient Descent
for i in range(num_iterations):
    # Calculate predictions
    y_pred = m * x + c
    
    # Compute the Gradients
    m_grad = (-2 / len(x)) * np.sum(x * (y-y_pred))
    c_grad = (-2 / len(x)) * np.sum(y - y_pred)
    
    # Update parameters
    m = m - learning_rate * m_grad
    c = c - learning_rate * c_grad 
    
    #Compute loss
    loss = np.mean((y - y_pred) ** 2)
    loss_history.append(loss)
    
    #print progress every 100 iterations
    if i % 100 == 0:
        print(f"Iteration {i}: m={m:.4f}, c={c:.4f}, loss={loss:.4f}")
    
```

```python
#print the final parameters
print(f"Final parameters: m={m:.4f}, c={c:.4f}")
```

```python
#Code the dataset and the fitted line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data points', color='blue')
plt.plot(x, m*x+c, label=f"Fitted Line: y = {m:.2f}x + {c:.2f}", color="red")
plt.title("Gradient Descent: Line Fitting")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Function to generate a 2D Gaussian distribution
plt.figure(figsize=(10,6))
plt.plot(range(num_iterations), loss_history, label="loss")
plt.title("Loass over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Mean Sqaured Error")
plt.legend()
plt.show()
```
